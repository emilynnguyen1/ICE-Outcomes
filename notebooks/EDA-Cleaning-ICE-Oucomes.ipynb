### Merging tables from 2023 - 2025 deportation data

```python
# Read each Excel file
arrests_2325 = pd.read_excel("../data/ice_release_2023to2025/ERO Admin Arrests_LESA-STU-FINAL Release.xlsx")
detainers_2325 = pd.read_excel("../data/ice_release_2023to2025/ERO Detainers_LESA-STU_FINAL Release.xlsx")
detentions_2325 = pd.read_excel("../data/ice_release_2023to2025/ICE Detentions_LESA-STU_FINAL Release.xlsx")

# Add source labels (optional but VERY useful for analysis later)
arrests_2325["source"] = "admin_arrests"
detainers_2325["source"] = "detainers"
detentions_2325["source"] = "detentions"

# Merge into one dataframe
ice_2023_2025_df = pd.concat(
    [arrests_2325, detainers_2325, detentions_2325],
    ignore_index=True
)

print(ice_2023_2025_df.shape)
```
